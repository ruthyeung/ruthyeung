```python
# Game: Rock paper scissors

# A function to play rock paper scissors with computer
# Requirement:
# - rules: paper > rock, rock > scissors, etc.
# - computer randomly choose one
# - player input a choice
# - win / lose / draw
# - 3 rounds, win 2 round then win the game
# - if draw, not count as 1 round
```

```python
# computer choose randomly
import random

choices = ['rock', 'paper', 'scissors']
computer_choice = random.choice(choices)

print(computer_choice)
```

```python
# alternative choices
choices = [0,1,2]
computer_choice = random.choice(choices)

print(computer_choice)
```

```python
rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

userScore = 0
compScore = 0


import random

for i in range(3):

  game_image = [rock, paper, scissors]
  user_choice = int(input("What do you choose? Type 0 for Rock, 1 for Paper or 2 for Scissors.\n"))
  if user_choice == 0:
    print(rock)
  if user_choice == 1:
    print(paper)
  if user_choice == 2:
    print(scissors)


  computer_chioce = random.randint(0,2)
  print("Computer chose:")
  print(game_image[computer_chioce])
  if user_choice >= 3 or user_choice < 0: 
    print("You typed an invalid number, you lose!") 
    
  if user_choice == 0 and computer_chioce == 2:
    print ("You win!")
  userScore += 1

  if user_choice == 0 and computer_chioce == 1:
    print ("You lose")
    compScore += 1
  if user_choice == 0 and computer_chioce == 0:
    print ("Its a draw")
  if user_choice == 1 and computer_chioce == 0:
    print ("You win!")
    userScore += 1
 
  if user_choice == 1 and computer_chioce == 1:
    print ("Its a draw")
  if user_choice == 1 and computer_chioce == 2:
    print ("You lose")
    compScore += 1

  if user_choice == 2 and computer_chioce == 0:
    print ("You lose")
  compScore += 1

  if user_choice == 2 and computer_chioce == 1:
    print ("You win!")
  userScore += 1

  if user_choice == 2 and computer_chioce == 2:
    print ("Its a draw")


if userScore > compScore:
    print("The user wins the game!")
elif compScore > userScore:
    print("The computer wins the game!")

```
